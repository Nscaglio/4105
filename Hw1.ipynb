```python
#Load the data set and seperate them into different lists
file_path = '/content/drive/My Drive/ENGR4105/Datasets/D3.csv'
data=pd.DataFrame(pd.read_csv(file_path))
x1=data.values[:, 0]
x2=data.values[:, 1]
x3=data.values[:, 2]
y=data.values[:, 3]
#find the lengths of the the different lists
m=len(y)
n1=len(x1)
n2=len(x2)
n3=len(x3)
#print the begining of the lists and their total lengths lengths
print('X1 = ', x1[: 5])
print('n1 = ', n1)
print('X2 = ', x2[: 5])
print('n2 = ', n2)
print('X3 = ', x3[: 5])
print('n3 = ', n3)
print('y = ', y[: 5])
print('m = ', m)
```

```python
#set X_0 to a vector of ones
X_0 = np.ones((m, 1))
#load X_1 through X_3 into arrays and stacking them with X_0
X_1 = x1.reshape(m, 1)
X_2 = x2.reshape(m, 1)
X_3 = x3.reshape(m, 1)
X1 = np.hstack((X_0, X_1))
X2 = np.hstack((X_0, X_2))
X3 = np.hstack((X_0, X_3))
#set theta
theta = np.zeros(2)
print(X3[: 5])
```

```python
def compute_cost(X, y, theta):
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    J = 1 / (2 * m) * np.sum(sqrErrors)
    return J
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)  # Number of training examples
    loss_history = np.zeros(iterations) #starts an array iterations long

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors)
        theta -= sum_delta          #sets this iterations theta
        loss_history[i] = compute_cost(X, y, theta)  # computes and saves the current iterations cost

    return theta, loss_history
```

```python
iterations = 1500
alpha = 0.1
```

```python
#calculations based on X1
theta1, cost1 = gradient_descent(X1, y, theta, alpha, iterations)
print("X1 based prediction = x *",theta1[1],"+",theta1[0])
print("end cost=",compute_cost(X1,y,theta1))
#Plot the X1 prediction compared to the scatter plot for X1
plt.scatter(X1[:, 1], y, color='red', marker='o', label='Training Data')
plt.plot(X1[:, 1], X1.dot(theta1), color='green', label='Linear Regression')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('x1')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
#plots the loss through the iterations
plt.plot(range(1, iterations + 1), cost1, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent')
plt.show()
```
